##Método Montecarlo para calcular $\pi$

Llamamos a la libreria para graficar

```julia
using PyPlot
```

Definimos las entradas de los vectores aleatorios

```julia
x_dentro=Float64[]
y_dentro=Float64[]
x_fuera=Float64[]
y_fuera=Float64[]
```

Realizamos el muestreo aleatorio del método montecarlo

```julia
for i in 1:1000000
    a=2*rand()-1
    b=2*rand()-1
    l=sqrt(a^2+b^2)
    if l<1
        push!(x_dentro,a)
        push!(y_dentro,b)
    else
        push!(x_fuera,a)
        push!(y_fuera,b)
    end
end
```

Graficamos las entradas

```julia
#Le damos formato a la gráfica
PyPlot.xlabel("eje x")
PyPlot.ylabel("eje y")
PyPlot.title("Montecarlo")
#Graficamos los puntos dentro 
PyPlot.plot(x_dentro,y_dentro,"ro", markersize=10)
#Graficamos los puntos afuera 
PyPlot.plot(x_fuera,y_fuera,"bo", markersize=10)
```

```julia
π=4*length(x_dentro)/(length(x_fuera)+length(x_dentro))
```
